```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('25Portfolios_C_t^p.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
```

```python
df.plot(y=['P1','P13','P25'], figsize=(10,6))
plt.legend(['Most illiquid portfolio','Medium illiquid portfolio','Least illiquid portfolio'])
plt.xlabel('Year', fontsize=16)
plt.ylabel(r'$c^p_t$', fontsize=16)
```

```python
df.plot(y=['P1','P25'], figsize=(10,6))
plt.legend(['Most illiquid portfolio','Least illiquid portfolio'])
plt.xlabel('Year', fontsize=16)
plt.ylabel(r'$c^p_t$', fontsize=16)
```
